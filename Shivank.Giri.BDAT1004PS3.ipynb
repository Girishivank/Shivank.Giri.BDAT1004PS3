Step 4. Year 2061? Do we really have data from this year? Create a function to fix it
and apply it

yes, we have data from 2061

```python
dataset.dtypes
def fix(x):
    if x.year > 2021:
        year = x.year - 100
    else:
        year = x.year
    return datetime.date(year, x.month, x.day)
dataset.Yr = dataset.Date.apply(fix)
data.head()
```
